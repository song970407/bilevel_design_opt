```python
num_x = 3
num_heaters = 5
r = 5
model_name = 'GAT'
data_ratio = 1.0

with open('design_opt_experiment/{}_{}/design_opt_experiment_result_{}_{}_{}.pkl'.format(num_x,
                                                                                         num_heaters,
                                                                                         model_name,
                                                                                         data_ratio,
                                                                                         r),
          'rb') as f:
    design_opt_experiment_result = pickle.load(f)
log = design_opt_experiment_result['design_opt_log']
```

```python
plt.style.use('default')
idx_list = [0, 10, 50, 100, log['best_idx']]

state_pos = design_opt_experiment_result['optimized_graph'].nodes['state'].data['pos'].detach().numpy()

for (i, idx) in enumerate(idx_list):
    fig, axes = plt.subplots(1, 1, figsize=(5, 5))
    if idx == 100:
        idx = 99
    action_pos = log['position_trajectory'][idx]
    axes.scatter(state_pos[:, 0], state_pos[:, 1], color='blue', label='Sensor', marker='o', s=200)
    axes.scatter(action_pos[:, 0], action_pos[:, 1], color='red', label='Heater', marker='s', s=100)
    loss = log['total_loss_trajectory'][idx]
    if idx < 100 and idx != log['best_idx']:
        grad = log['position_trajectory'][idx+1] - log['position_trajectory'][idx]
        for k in range(grad.shape[0]):
            plt.arrow(action_pos[k, 0], action_pos[k, 1], grad[k, 0], grad[k, 1], color='red', linewidth=3, head_width=0.05, head_length=0.05)
    if i == 0:
        axes.set_title('Initial Layout', fontsize=25)
    elif idx!=log['best_idx']:
        if idx == 99:
            axes.set_title('{}th Step'.format(idx+1), fontsize=25)
        else:
            axes.set_title('{}th Step'.format(idx), fontsize=25)
    else:
        axes.set_title('Best at {}th Step'.format(log['best_idx']), fontsize=25)
    axes.set_xlim([-1, 1])
    axes.set_ylim([-1, 1])
    
    tick = [-1.0, '', -0.5, '', 0.0, '', 0.5, '', 1.0]
    axes.set_xticklabels(tick, fontsize=12)
    axes.set_yticklabels(tick, fontsize=12)
    
    axes.set_xlabel('Obj: {}'.format(np.round(loss, 5)), fontsize=25)
    if i == 0:
        axes.legend(fontsize=15, loc=1)
    fig.tight_layout()
    fig.show()
```

```python
num_x = 3
num_heaters = 5
r = 7
model_name = 'ICGAT'
data_ratio = 1.0
```

```python
with open('design_opt_experiment/{}_{}/design_opt_experiment_result_{}_{}_{}.pkl'.format(num_x,
                                                                                         num_heaters,
                                                                                         model_name,
                                                                                         data_ratio,
                                                                                         r),
          'rb') as f:
    design_opt_experiment_result = pickle.load(f)
log = design_opt_experiment_result['design_opt_log']
u_trajectory = log['us_trajectory'][log['best_idx']].reshape((5, 32, 24))

with open('design_opt_experiment/optimal/{}_{}/optimal_experiment_result_{}_{}_{}.pkl'.format(num_x,
                                                                                              num_heaters,
                                                                                              model_name,
                                                                                              data_ratio,
                                                                                              r),
          'rb') as f:
    optimal_experiment_result = pickle.load(f)
true_u_trajectory_list = optimal_experiment_result['u_trajectory_list']
true_u_trajectory_list = np.stack(true_u_trajectory_list, axis=1)


for i in range(32):
    fig, axes = plt.subplots(1, 5, figsize=(10, 2))
    axes_flatten = axes.flatten()
    for j in range(5):
        axes_flatten[j].plot(u_trajectory[j, i] * 10, label='Expected')
        axes_flatten[j].plot(true_u_trajectory_list[:, i, j], label='True')
        axes_flatten[j].legend()
    fig.show()
```
